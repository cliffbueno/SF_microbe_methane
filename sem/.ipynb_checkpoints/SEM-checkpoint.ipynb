## Setup

```r
# Libraries
suppressWarnings(library(tidyverse))
suppressWarnings(library(cowplot))
suppressMessages(suppressWarnings(library(lavaan)))
`%notin%` <- Negate(`%in%`)
```

```r
# Data - need merged metadata and guilds
Silva_gc <- read.table("../guild_analysis/Silva_OTU_Guild_abundT_counts.txt", sep = '\t') %>%
    rownames_to_column(var = "Sample")

metaDB <- read.table("../data/meta/SF_sal_meta_FIX3.5.txt", sep = "\t", header = TRUE) %>%
    left_join(., Silva_gc, by = "Sample")
names(metaDB)
```

## SEM

```r
# Try basic SEM with lavaan
myModel <- '
   # regressions
     CH4_ac ~ CO2_mg_m2_h + H2O_FPS + SRB + SRB_syn + Salinity.x
     MOB_IIa ~ Bulk_dens + NOB + AOA + Salinity.x
     CH4_ug_m2_h ~ CH4_ac + MOB_IIa
'
fit <- sem(model = myModel,
           data  = metaDB)
summary(fit)
```
