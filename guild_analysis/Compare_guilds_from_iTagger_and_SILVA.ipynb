# Compare guilds from iTagger and SILVA

```r
library(ggplot2)
suppressMessages(library(gridExtra))
# library(corrgram)
# library(corrplot)
# suppressMessages(library(Hmisc))
```

# 1) import data

```r
# Import aggregated guild count tables
iTagger_gc <- read.table("../data/iTaggerOTUs/guild_data/OTU_Guild_abundT_counts.txt", sep='\t')
names(iTagger_gc) <- paste0(names(iTagger_gc),"_i")
# iTagger_gc
```

```r
# Silva
Silva_gc <- read.table("Silva_OTU_Guild_abundT_counts.txt", sep='\t')
names(Silva_gc) <- paste0(names(Silva_gc),"_s")
# Silva_gc
```

```r
# combine
Combined_guilds = data.frame(iTagger_gc, Silva_gc)
```

# 2) define plots
- the ugly way, a should be function over list
- log version
