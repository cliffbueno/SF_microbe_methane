```r
# Import TreeSAPP
ts_table <- read.delim("../data/shotgun/TreeSapp_v0.5_parsed_tables/TS_cleaned_guild_abund_table.txt") %>%
    select(-OTU, -Guild) %>%
    t() %>%
    as.data.frame() %>%
    mutate(p_amo_tot = AmoA.Ic + amoA.AOB + amoA.AOA + p.amo.unk + MOB.III + MOB.IIb + MOB.IIa + MOB.Ic + MOB.Ib + MOB.Ia) %>%
    mutate(amoA_tot = AmoA.Ic + amoA.AOB + amoA.AOA) %>%
    mutate(pmoA_tot = MOB.III + MOB.IIb + MOB.IIa + MOB.Ic + MOB.Ib + MOB.Ia) %>%
    mutate(Total_pmoA_amoA = p_amo_tot) %>% # Rename for Figure
    mutate(pmoA_MOB = pmoA_tot) %>% # Rename for Figure
    mutate(amoA_AOA = amoA.AOA) %>% # Rename for Figure
    mutate(amoA_AOB = amoA.AOB) %>% # Rename for Figure
    filter(rownames(.) %in% rownames(ko_table))
```

```r
# Import BGC data

# From Figure1.ipynb:

# Colors
site_colors <- read.table("../data/colors/Sal_siteColors_testR.txt", sep = '\t', header = T, row.names = 1) %>%
    set_names(c('color','Location'))
# site_colors

# Metadata - filter to those with MG_names
metaDB <- read.table("../data/meta/SF_sal_meta_FIX3.5.txt", sep = "\t", header = TRUE) %>%
    mutate(Location = factor(Location, levels = rownames(site_colors))) %>%
    mutate(CH4_logn1 = log10(CH4_ug_m2_h - (1.05*min(CH4_ug_m2_h)))) %>%
    mutate(CO2_logn1 = log10(CO2_mg_m2_h - (1.05*min(CO2_mg_m2_h)))) %>%
    filter(MG_name != "NA")
```

```r
# Merge with TreeSAPP data
nrow(metaDB)
nrow(ts_table)
`%notin%` <- Negate(`%in%`)
mismatch1 <- metaDB %>%
    filter(Sample %notin% rownames(ts_table))
mismatch1 # 4 with MG names not in ts_table
mismatch2 <- ts_table %>%
    filter(rownames(ts_table) %notin% metaDB$Sample)
mismatch2 # 3 redone MG not in metaDB
# Removing those we should have a match of 130 samples
```

```r
metaDB <- metaDB %>%
    filter(Sample %in% rownames(ts_table)) %>%
    arrange(Sample)
ts_table <- ts_table %>%
    filter(rownames(.) %in% metaDB$Sample) %>%
    rownames_to_column(var = "Sample") %>%
    arrange(Sample)
nrow(metaDB)
nrow(ts_table)
sum(ts_table$Sample != metaDB$Sample)
dim(metaDB)
metaDB <- metaDB %>%
    left_join(., ts_table, by = "Sample")
dim(metaDB)
```

```r
metaDB$pmoA_MOB
```

```r
# Remove NA
metaDB <- metaDB %>%
    filter(is.na(pmoA_MOB) == FALSE)
nrow(metaDB)
```

### Stats

```r
# Stats log CH4 versus pmoA_MOB, amoA_AOA, amoA_AOB (use)
print('Log CH4, pmoA_MOB - Linear')
m1 <- lm(CH4_logn1 ~ pmoA_MOB, data = metaDB)
round(summary(m1)$r.squared, digits = 2) # R Squared
round(summary(m1)$coefficients[2,4], digits = 4) # P-value
print('Log CH4, pmoA_MOB - Polynomial')
m2 <- lm(CH4_logn1 ~ poly(pmoA_MOB, 2), data = metaDB)
round(summary(m2)$r.squared, digits = 2) # R Squared
round(summary(m2)$coefficients[2,4], digits = 4) # P-value
print('Compare Models')
anova(m1, m2)

print('Log CH4, amoA_AOA - Linear')
m3 <- lm(CH4_logn1 ~ amoA_AOA, data = metaDB)
round(summary(m3)$r.squared, digits = 2) # R Squared
round(summary(m3)$coefficients[2,4], digits = 4) # P-value
print('Log CH4, amoA_AOA - Polynomial')
m4 <- lm(CH4_logn1 ~ poly(amoA_AOA, 2), data = metaDB)
round(summary(m4)$r.squared, digits = 2) # R Squared
round(summary(m4)$coefficients[2,4], digits = 4) # P-value
print('Compare Models')
anova(m3, m4)

print('Log CH4, amoA_AOB - Linear')
m5 <- lm(CH4_logn1 ~ amoA_AOB, data = metaDB)
round(summary(m5)$r.squared, digits = 2) # R Squared
round(summary(m5)$coefficients[2,4], digits = 4) # P-value
print('Log CH4, amoA_AOB - Polynomial')
m6 <- lm(CH4_logn1 ~ poly(amoA_AOB, 2), data = metaDB)
round(summary(m6)$r.squared, digits = 2) # R Squared
round(summary(m6)$coefficients[2,4], digits = 4) # P-value
print('Compare Models')
anova(m1, m2)

# Stats total pmoA versus total amoA
print('pmoA, amoA - Linear')
m7 <- lm(pmoA_tot ~ amoA_tot, data = metaDB)
round(summary(m7)$r.squared, digits = 2) # R Squared
round(summary(m7)$coefficients[2,4], digits = 3) # P-value
print('pmoA, amoA - Polynomial')
m8 <- lm(pmoA_tot ~ poly(amoA_tot, 2), data = metaDB)
round(summary(m8)$r.squared, digits = 2) # R Squared
round(summary(m8)$coefficients[2,4], digits = 3) # P-value
print('Compare Models')
anova(m7, m8)
```

```r
# Correlation
cor.test(metaDB$pmoA_MOB, metaDB$CH4_logn1, method = "pearson")
cor.test(metaDB$pmoA_MOB, metaDB$CH4_logn1, method = "spearman")
cor.test(metaDB$pmoA_MOB, metaDB$CH4_logn1, method = "kendall")
```
